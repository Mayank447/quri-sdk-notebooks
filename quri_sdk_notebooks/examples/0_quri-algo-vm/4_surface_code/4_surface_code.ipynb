# Surface Code and Quantum Error Correction

## Basic Error Correction
Quantum error correction is essential for protecting quantum information from errors due to decoherence and other quantum noise. The basic idea is to encode a logical qubit into a larger number of physical qubits in such a way that errors can be detected and corrected.

In quantum error correction, the code distance is a crucial parameter that measures the robustness of a quantum error-correcting code. It is defined as the minimum number of qubits that need to be altered to transform one valid codeword into another. Essentially, it represents the minimum weight of an error that can cause a logical error in the code.

Here's a more detailed breakdown:

Quantum Error-Correcting Code (QECC): This is a method used to protect quantum information from errors due to decoherence and other quantum noise. It involves encoding the quantum information into a larger Hilbert space using additional qubits.

Code Distance (d): The code distance is the smallest number of qubits that must be flipped (or otherwise altered) to change one valid encoded state (codeword) into another. It is a measure of the code's ability to detect and correct errors. A higher code distance means the code can detect and correct more errors.

Error Detection and Correction: For a code with distance ($d$), it can detect up to $(d-1)$ errors and correct up to $(\left\lfloor \frac{d-1}{2} \right\rfloor)$ errors. This means that if the code distance is 5, the code can detect up to 4 errors and correct up to 2 errors.

Logical Qubits and Physical Qubits: In a QECC, logical qubits are the qubits that store the actual quantum information, while physical qubits are the qubits used to encode the logical qubits. The code distance is related to the number of physical qubits used to protect each logical qubit.

For example, in the case of the well-known Shor code, which encodes one logical qubit into nine physical qubits, the code distance is 3. This means that it can correct any single-qubit error.

Understanding code distance is essential for designing robust quantum error-correcting codes that can effectively protect quantum information against errors in quantum computing systems.

## Syndrome Measurements
Syndrome measurements are used to detect errors without collapsing the quantum state. By measuring specific operators (syndromes), we can determine the type and location of errors. For example, in the case of bit-flip errors, we measure the parity of certain qubits.

## Surface Code
The surface code is a type of topological quantum error-correcting code that arranges qubits on a 2D lattice. It is highly robust against errors and has a high threshold for fault-tolerant quantum computation. The surface code uses two types of stabilizers: star operators and plaquette operators.

### Star Operators
Star operators are defined as:
$$ A_s = \prod_{i \in s} X_i $$
where \( X_i \) is the Pauli-X operator acting on qubit \( i \).

### Plaquette Operators
Plaquette operators are defined as:
$$ B_p = \prod_{i \in p} Z_i $$
where \( Z_i \) is the Pauli-Z operator acting on qubit \( i \).

## Solovay-Kitaev Decomposition
The Solovay-Kitaev theorem states that any quantum gate can be approximated to arbitrary accuracy using a finite set of universal gates. The Solovay-Kitaev algorithm provides a method to decompose a given unitary operation into a sequence of gates from this universal set.

## Magic State Distillation
Magic state distillation is a process used to produce high-fidelity quantum states (magic states) that are necessary for implementing certain quantum gates, such as the T-gate, in a fault-tolerant manner. The process involves preparing multiple noisy copies of a magic state and then using error correction techniques to distill a smaller number of higher-fidelity magic states.

In summary, the surface code and related techniques like syndrome measurements, Solovay-Kitaev decomposition, and magic state distillation are crucial components of fault-tolerant quantum computation. They enable the reliable execution of quantum algorithms despite the presence of errors and noise.

The **rotated planar surface code** is a variant of the **surface code**, which is one of the most promising quantum error-correcting codes for practical implementations of fault-tolerant quantum computation. The rotated version is designed to optimize the layout of physical qubits and simplify the implementation while maintaining the error-correcting capabilities of the standard surface code.

---

### **Standard Surface Code Recap**
The standard surface code is defined on a 2D lattice of physical qubits, with stabilizer measurements performed using neighboring qubits to detect errors. It encodes one logical qubit into a grid of physical qubits, and its code distance \( d \) is equal to the linear size of the lattice (e.g., a \( d \times d \) lattice has distance \( d \)).

---

### **Rotated Planar Surface Code**
The rotated planar surface code modifies the standard surface code by rotating the lattice by 45 degrees. This rotation changes the connectivity and layout of the qubits, leading to several advantages:

1. **Compact Layout**:
   - In the rotated version, the lattice is effectively "squashed" into a diamond shape, reducing the number of physical qubits required to achieve a given code distance \( d \).
   - For a code distance \( d \), the rotated planar surface code requires approximately \( d^2 \) physical qubits, compared to \( (2d-1)^2 \) for the standard surface code. This makes it more resource-efficient.

2. **Simplified Connectivity**:
   - The rotated lattice simplifies the connectivity requirements for stabilizer measurements. Each stabilizer (X or Z) involves only four neighboring qubits, arranged in a square, which is easier to implement in hardware architectures like superconducting qubits or photonic qubits.

3. **Error Correction Performance**:
   - The rotated planar surface code retains the same error-correcting capabilities as the standard surface code. It can correct up to \( t = \lfloor \frac{d-1}{2} \rfloor \) errors and has a code distance \( d \).

---

### **Structure of the Rotated Planar Surface Code**
- **Qubit Arrangement**:
  - Physical qubits are placed on the vertices of a rotated square lattice, forming a diamond-shaped grid.
  - Data qubits (used to encode the logical state) and measure qubits (used for stabilizer measurements) are interleaved in a checkerboard pattern.

- **Stabilizers**:
  - Stabilizers are defined on plaquettes (squares) of the lattice.
  - X-type stabilizers measure the parity of X-operators on the four qubits around a plaquette.
  - Z-type stabilizers measure the parity of Z-operators on the four qubits around a plaquette.

- **Boundary Conditions**:
  - The boundaries of the rotated lattice are modified to ensure that the code can still detect and correct errors effectively. For example, some stabilizers may involve only three qubits at the edges.

---

### **Advantages of the Rotated Planar Surface Code**
1. **Reduced Qubit Overhead**:
   - The rotated layout requires fewer physical qubits to achieve the same code distance, making it more scalable for near-term quantum devices.

2. **Easier Hardware Implementation**:
   - The simplified connectivity and compact layout make it easier to map the code onto physical hardware with limited qubit connectivity.

3. **Maintained Error Correction**:
   - Despite the reduced resource requirements, the rotated planar surface code retains the same error-correcting capabilities as the standard surface code.

---

### **Example: Rotated Planar Surface Code with \( d = 3 \)**
- The lattice consists of \( d^2 = 9 \) physical qubits arranged in a diamond shape.
- Stabilizers are measured on the plaquettes, with X-type and Z-type stabilizers alternating in a checkerboard pattern.
- The code can correct up to \( t = 1 \) error (since \( t = \lfloor \frac{3-1}{2} \rfloor = 1 \)).

---

### **Applications**
The rotated planar surface code is particularly well-suited for near-term quantum devices, where qubit counts and connectivity are limited. It is a leading candidate for implementing fault-tolerant quantum computation in hardware platforms like superconducting qubits, trapped ions, and photonic qubits.

---

In summary, the **rotated planar surface code** is a resource-efficient variant of the surface code that simplifies the physical layout and connectivity while maintaining strong error-correcting capabilities. Its compact design makes it a practical choice for realizing fault-tolerant quantum computation in real-world systems.

## Problem setup

We will set up the problem that we will simulate. This is going to be the Heisenberg model without any anisotropies. First let us set it up and then see how the circuit depth depends on the number of Trotter steps chosen for the time-evolution.

```python
from functools import cache

from quri_algo.circuit.time_evolution.trotter_time_evo import TrotterTimeEvolutionCircuitFactory
from quri_algo.problem import QubitHamiltonianInput
from quri_parts.circuit import NonParametricQuantumCircuit
from quri_parts.core.operator import Operator, pauli_label

def get_operator(qubit_count: int) -> Operator:
    op = Operator()
    for q in range(qubit_count):
        for p in ['X', 'Y', 'Z']:
            op.add_term(pauli_label(p + str(q) + ' ' + p + str((q+1)%qubit_count)), 1.0)
    
    return op

class HeisenbergTimeEvolutionCircuitFactory:
    """Time-evolution circuit factory
    
    This class wraps TrotterTimeEvolutionCircuitFactory and allows creating circuits specifically for the Heisenberg model by calling it with the qubit count and evolution time"""
    # Return a time-evolution circuit factory and cache it
    @cache
    def get_circuit_factory(self, qubit_count:int, n_trotter: int) -> TrotterTimeEvolutionCircuitFactory:
        operator = get_operator(qubit_count)
        hamiltonian_input = QubitHamiltonianInput(qubit_count, operator)

        return TrotterTimeEvolutionCircuitFactory(hamiltonian_input, n_trotter)

    def __call__(
        self, evolution_time: float, qubit_count: int, n_trotter: int
    ) -> NonParametricQuantumCircuit:
        circuit_factory = self.get_circuit_factory(qubit_count, n_trotter)
        return circuit_factory(evolution_time=evolution_time)
```

```python
from quri_algo.core.cost_functions import HilbertSchmidtTest
from quri_algo.circuit.time_evolution.exact_unitary import ExactUnitaryTimeEvolutionCircuitFactory
from quri_parts.qulacs.estimator import create_qulacs_vector_estimator

N_QUBITS = 8
EVOLUTION_TIME = 1.0
TROTTER_STEPS = [1, 3, 10, 30, 100]
SK_EXPONENTS = [-2,-3,-4]


estimator = create_qulacs_vector_estimator()
hs_test = HilbertSchmidtTest(estimator)

circuit_factory = HeisenbergTimeEvolutionCircuitFactory()
circuit_factory_exact = ExactUnitaryTimeEvolutionCircuitFactory(QubitHamiltonianInput(N_QUBITS,get_operator(N_QUBITS)))

fidelities = []
hs_cost = []
transpiled_depth = []

for n_t in TROTTER_STEPS:
    time_evo_exact = circuit_factory_exact(EVOLUTION_TIME)
    depth = {}
    fidelity = {}
    time_evo_trotter = circuit_factory(EVOLUTION_TIME, N_QUBITS, n_t)
    for e in SK_EXPONENTS:
        vm = get_vm(qubit_count=N_QUBITS, delta_sk=10**e)
        time_evo_trotter_transpiled = vm.transpile(time_evo_trotter)
        analysis = vm.analyze(time_evo_trotter)
        fidelity[e] = analysis.fidelity
        depth[e] = vm.analyze(time_evo_trotter).depth
    time_evo_trotter = circuit_factory(EVOLUTION_TIME, N_QUBITS, n_t)
    hs_cost.append(hs_test(time_evo_exact,time_evo_trotter))
    depth[0] = time_evo_trotter.depth
    fidelities.append(fidelity)
    transpiled_depth.append(depth)
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

color = 'tab:red'
linestyles = {
    0: '-',
    -2: ':',
    -3: '-.',
    -4: '--',
}
ax1.set_xlabel('Number of Trotter Steps')
ax1.set_ylabel('Clifford+T Circuit Depth', color=color)
for e in SK_EXPONENTS:
    ax1.plot(TROTTER_STEPS, [depth[e] for depth in transpiled_depth], label=f'delta-sk: {10**e}', color=color, linestyle=linestyles[e])
ax1.plot(TROTTER_STEPS, [depth[0] for depth in transpiled_depth], label=f'logical circuit depth', color=color, linestyle=linestyles[0])
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend()

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Fidelity and HS-cost', color=color)
for e in SK_EXPONENTS:
    ax2.plot(TROTTER_STEPS, [fidelity[e] for fidelity in fidelities], label=f'delta-sk: {10**e}', color=color, linestyle=linestyles[e])
ax2.plot(TROTTER_STEPS, [hs.value.real for hs in hs_cost], label=f'Hilbert-Schmidt cost', color=color, linestyle='-')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Fidelities, Transpiled Depth, and Hilbert-Schmidt Cost vs Trotter Steps')
plt.legend()
plt.show()
```
